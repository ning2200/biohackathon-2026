# PCOS Dataset Exploration

A comprehensive exploration of the main PCOS dataset to understand its structure, content, and summary statistics.

## 1. Import Required Libraries

Import necessary libraries such as pandas, numpy, and matplotlib for data loading and exploration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

## 2. Load the Dataset

Read the main dataset from the Excel file.

```python
# Load the dataset from Excel file
dataset_path = 'Dataset/(Main_Dataset)_PCOS_data_without_infertility.xlsx'
df = pd.read_excel(dataset_path)

print(f"Dataset loaded successfully!")
print(f"File path: {dataset_path}")
```

## 3. Inspect Dataset Structure

Examine the shape, columns, and index of the dataset.

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names and Info:")
df.info()
print("\nDataset Columns:")
print(df.columns.tolist())
```

## 4. Display First Few Rows

View the first and last rows of the dataset.

```python
print("First 5 rows of the dataset:")
print(df.head())
print("\n" + "="*80 + "\n")
print("Last 5 rows of the dataset:")
print(df.tail())
```

## 5. Check Data Types and Missing Values

Verify data types of each column and identify missing values.

```python
print("Data Types:")
print(df.dtypes)
print("\n" + "="*80 + "\n")
print("Missing Values:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0] if (missing_values > 0).any() else "No missing values found!")
print("\n" + "="*80 + "\n")
print("Missing Values Percentage:")
missing_percentage = (df.isnull().sum() / len(df) * 100)
print(missing_percentage[missing_percentage > 0] if (missing_percentage > 0).any() else "No missing values found!")
```

## 6. Get Dataset Summary Statistics

Generate descriptive statistics for numerical and categorical columns.

```python
print("Summary Statistics for Numerical Columns:")
print(df.describe())
print("\n" + "="*80 + "\n")
print("Summary Statistics for Categorical Columns:")
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    for col in categorical_cols:
        print(f"\n{col}:")
        print(df[col].value_counts())
else:
    print("No categorical columns found!")
```
